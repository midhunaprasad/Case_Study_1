# 1. Reading the dataset to python environment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
sns.set()
filename='cars_data.csv'
df=pd.read_csv(filename)
df
```

# 2. Checking for the null values present in the dataset.

```python
df.info()
```

<h5>From the above output we can see that there are 10000 rows and 20 columns in the dataset and there are no null values in the data set.</h5>

# 3. Bar graph of male vs female buyers participated in the sales.

```python
df['Buyer Gender'].value_counts()
```

<h5>From the above output we can see that there is a difference of 52 in the count of male and female participants in the sales, i.e the number of female participants are more by 52 than the number of male participants.</h5>

```python
sns.set(font_scale=1.4)
df['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0,color=['red','blue'])
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title('Bar Graph of Male Vs Female Buyers Participated in the Sales')
```

<h5>From the above output of the bar chart we can clearly see that the Male and Female participation in the sale is not equal but the female participation is slightly greater than the male participation</h5>

# 4. The top 5 cars based on their sales price.

```python
top_sale_price= df.sort_values(by=['Sale Price']).tail()
top_cars_df = [top_sale_price["Make"]+" "+top_sale_price["Model"],top_sale_price['Sale Price']]
headers = ["Car Name", "Sale Price"]
top_cars = pd.concat(top_cars_df, axis=1, keys=headers)
```

```python
top_cars
```

<h5>The above are the top 5 cars based on their sales price i.e the top 5 cars with the highest sales price.</h5>

# 5. The least 5 cars based on their Resell price.

```python
least_resell_price= df.sort_values(by=['Resell Price']).head()
least_cars_df = [least_resell_price["Make"]+" "+least_resell_price["Model"],least_resell_price['Resell Price']]
headers = ["Car Name", "Resell Price"]
least_cars = pd.concat(least_cars_df, axis=1, keys=headers)
```

```python
least_cars
```

<h5>The above are the least 5 cars based on their resell price i.e the least 5 cars with the lowest resell price.</h5>
